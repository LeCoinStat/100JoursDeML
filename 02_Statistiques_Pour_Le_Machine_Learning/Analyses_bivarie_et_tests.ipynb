

## Test de corrlation de Pearson

### Hypothses :
- H0(hypothse nulle) : Il n'y a pas de corrlation **linaire** entre les deux variables.
- H1(hypothse alternative) : Il y a une corrlation **linaire** entre les deux variables.

### Conditions d'utilisation :
- Les deux variables sont quantitatives et mesures sur une chelle continue.
- La relation entre les deux variables est linaire
- Les deux variables suivent une **distribution normale**.


### Interprtation :
Le coefficient de corrlation de pearson varie entre -1 et 1. 

- Si le coefficient est proche de 1 par exemple (0.6), il y a une corrlation positive entre les variables.
- Si le coefficient est proche de -1 (par exemple -0.6), il y a une corrlation ngative entre les variables.
- Si le coefficient est proche de 0, il n'y a pas de corrlation linaire entre les variables.


Vous pouvez aussi regarder la pvaleur du test:

- Si p < 0.05, on rejette H0 et on conclut qu'il y a une corrlation significative entre les deux variables.
- Si p >= 0.05, on ne peut pas rejeter H0 et on conclut qu'il n'y a pas suffisamment de preuves pour dire qu'il y a une corrlation significative entre les deux variables.



```python
# Analyse de la normalit grce au test de shapiro
from scipy.stats import shapiro, pearsonr, spearmanr, kendalltau
# Variables quantitatives d'intrt
var1 = 'sepal_length'
var2 = 'petal_length'

# Analyse graphique de la normalit des deux variables
for var in [var1, var2]:
    plt.figure(figsize=(8, 6))
    sns.histplot(iris[var], kde=True)
    plt.xlabel(var)
    plt.ylabel('Frquence')
    plt.title(f"Histogramme et courbe de densit pour {var}")
    plt.show()

# Vrifier la normalit  l'aide du test de Shapiro-Wilk
for var in [var1, var2]:
    _, p_value = shapiro(iris[var])
    print(f"P-valeur de la normalit (Shapiro-Wilk) pour {var}:", p_value)

```

```python
# Calculer les coefficients de corrlation de Spearman, Pearson et Kendall
pearsonr(iris[var1], iris[var2])
```

# Analyse de la liaion entre deux variables qualitatives

Dans cette partie, nous allons explorer diffrents tests **d'association** pour tudier la relation entre deux variables qualitatives. Nous utiliserons les tests du chi-carr (chi2) et de Cramr-V. Nous utiliserons le jeu de donnes "Titanic" populaire pour dmontrer l'application de ces tests en Python.


```python
#Analyse graphique entre sex et class de titanic 
titanic=sns.load_dataset("titanic")
# Crer un graphique de la relation entre 'sex' et 'class' dans Titanic
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='class', hue='sex')
plt.xlabel('Classe')
plt.ylabel('Nombre de passagers')
plt.title('Relation entre le sexe et la classe dans le Titanic')
plt.show()
```


## 1. Test du chi-carr (chi2)

### Hypothses
- H0 (hypothse nulle): Il n'y a pas d'association entre les deux variables qualitatives.
- H1 (hypothse alternative): Il y a une association entre les deux variables qualitatives.

### Conditions d'utilisation
- Les deux variables sont qualitatives (catgoriques).
- Les effectifs attendus sont suprieurs  5 pour chaque croissement de variable

### Interprtation
Si la p-valeur est infrieure au seuil de significativit (gnralement 0,05), on rejette l'hypothse nulle et on conclut qu'il y a une association entre les deux variables qualitatives.



```python
# Calcul de chi deux sur la base titanic
from scipy.stats import chi2_contingency
# Crer un tableau de contingence
contingency_table = pd.crosstab(titanic['sex'], titanic['class'])
contingency_table
```

```python

# Effectuer le test du chi-carr
chi2_stat, p_value, dof, expected_freq = chi2_contingency(contingency_table)

# Afficher les rsultats
print("Statistique du chi-carr :", chi2_stat)
print("P-valeur :", p_value)
print("Degrs de libert :", dof)
print("Frquences attendues :\n", expected_freq)
```


## 2. Coefficient de Cramr-V

Le coefficient de Cramr-V mesure la force de l'association entre deux variables qualitatives. Il varie entre 0 (pas d'association) et 1 (association parfaite).

### Conditions d'utilisation
- Les deux variables sont qualitatives (catgoriques).
- Le test du chi-carr a dj t effectu et a montr une association significative entre les deux variables.

### Interprtation
Plus la valeur de **Cramr-V est proche de 1 (suprieur  0.60)**, plus l'association entre les deux variables est forte.

```python
# Calculer le coefficient de Cramr-V
n = contingency_table.sum().sum()
min_dim = min(contingency_table.shape) - 1
cramer_v = np.sqrt(chi2_stat / (n * min_dim))

# Afficher le coefficient de Cramr-V
print("Coefficient de Cramr-V :", cramer_v)
```

## Coefficient T de Tschuprow

Le coefficient T de Tschuprow est une mesure de liaison entre deux variables qualitatives. Il est bas sur le chi-carr () et prend en compte la taille de l'chantillon et le nombre de catgories dans chaque variable.

### Hypothses
- H0 (hypothse nulle): Il n'y a pas de relation entre les deux variables qualitatives.
- H1 (hypothse alternative): Il y a une relation entre les deux variables qualitatives.

### Conditions d'utilisation
- Les variables sont qualitatives.
- Les observations sont indpendantes.
- Les effectifs des cellules du tableau de contingence sont suffisants (gnralement, chaque cellule doit avoir un effectif d'au moins 5).

### Interprtation
Le coefficient T de Tschuprow varie entre 0 et 1. Une valeur proche de 0 indique une faible liaison entre les deux variables, tandis qu'une valeur proche de 1 indique une liaison forte. La force de la liaison dpend du contexte et de la discipline, mais en gnral, une valeur suprieure  0,3 est considre comme une liaison modre et une valeur **suprieure  0,5 est considre comme une liaison forte**.

```python
# Calculer le coefficient T de Tschuprow
n = titanic.shape[0]
min_dim = min(contingency_table.shape) - 1
# Effectuer le test du chi-carr
chi2, p_value, _, _ = st.chi2_contingency(contingency_table)
tschuprow_t = np.sqrt(chi2 / (n * min_dim))

print("Coefficient T de Tschuprow :", tschuprow_t)
print("P-valeur :", p_value)
```

# Analyse de la liaison entre une variable qualitative et une variable quantitative

Dans cette, nous allons explorer diffrents tests pour tudier la liaison entre une variable qualitative et une variable quantitative. Nous utiliserons les tests de Student (t-test), ANOVA, Mann-Whitney, Kruskal-Walli. Nous utiliserons le jeu de donnes "Iris" populaire pour dmontrer l'application de ces tests en Python.

## Analyse graphique

```python

# Crer un boxplot de la relation entre 'species' et 'sepal_length' dans Iris
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.xlabel('Espces')
plt.ylabel('Longueur du spale')
plt.title('Relation entre les espces et la longueur du spale dans Iris')
plt.show()
```



##  Test de Student (t-test)

### Hypothses
- H0 (hypothse nulle): Les **moyennes** des deux groupes sont gales.
- H1 (hypothse alternative): Les moyennes des deux groupes sont diffrentes.

### Conditions d'utilisation
- La variable quantitative est **normalement** distribue dans chaque groupe.
- **Les variances des deux groupes sont gales**.
- Les observations sont indpendantes.

### Interprtation
Si la p-valeur est infrieure au seuil de significativit (gnralement 0,05), on rejette l'hypothse nulle et on conclut que les moyennes des deux groupes sont significativement diffrentes.



```python
from scipy.stats import ttest_ind, f_oneway, mannwhitneyu, kruskal, wilcoxon

# Charger le jeu de donnes Iris
iris = sns.load_dataset('iris')

# Sparer les donnes en deux groupes pour le test de Student et le test de Mann-Whitney
setosa = iris[iris['species'] == 'setosa']['sepal_length']
virginica = iris[iris['species'] == 'virginica']['sepal_length']


```

```python
## Vrification de la normalit

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, ttest_ind, f_oneway, mannwhitneyu, kruskal

# Charger le jeu de donnes Iris
iris = sns.load_dataset('iris')

# Sparer les donnes en groupes en fonction de l'espce
setosa = iris[iris['species'] == 'setosa']['sepal_length']
virginica = iris[iris['species'] == 'virginica']['sepal_length']
versicolor = iris[iris['species'] == 'versicolor']['sepal_length']

# Vrifier la normalit  l'aide du test de Shapiro-Wilk
_, setosa_p_value = shapiro(setosa)
_, virginica_p_value = shapiro(virginica)
_, versicolor_p_value = shapiro(versicolor)

print("P-valeur de la normalit (Shapiro-Wilk) - Setosa:", setosa_p_value)
print("P-valeur de la normalit (Shapiro-Wilk) - Virginica:", virginica_p_value)
print("P-valeur de la normalit (Shapiro-Wilk) - Versicolor:", versicolor_p_value)

# Histogrammes pour vrifier visuellement la normalit
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.hist(setosa, bins=10, color='blue', alpha=0.7)
plt.title("Setosa - Sepal Length")
plt.subplot(132)
plt.hist(virginica, bins=10, color='green', alpha=0.7)
plt.title("Virginica - Sepal Length")
plt.subplot(133)
plt.hist(versicolor, bins=10, color='red', alpha=0.7)
plt.title("Versicolor - Sepal Length")
plt.show()

# Botes  moustaches pour vrifier visuellement la normalit
plt.figure(figsize=(12, 6))
plt.boxplot([setosa, virginica, versicolor], labels=['Setosa', 'Virginica', 'Versicolor'])
plt.title("Boxplots - Sepal Length")
plt.show()
```

```python

# Test de Student (t-test)
t_stat, t_p_value = ttest_ind(setosa, virginica)
print("Statistique t de Student :", t_stat)
print("P-valeur (t-test) :", t_p_value)
```

##  Test de Mann-Whitney

### Hypothses
- H0 (hypothse nulle): Les distributions des deux groupes sont gales.
- H1 (hypothse alternative): Les distributions des deux groupes sont diffrentes.

### Conditions d'utilisation
- Les observations sont indpendantes.
- Les donnes peuvent tre ordonnes (au moins ordinale).


```python
# Test de Mann-Whitney
mann_whitney_stat, mann_whitney_p_value = mannwhitneyu(setosa, virginica)
print("Statistique U de Mann-Whitney :", mann_whitney_stat)
print("P-valeur (Mann-Whitney) :", mann_whitney_p_value)
```

## ANOVA (Analysis of Variance)

### Hypothses
- H0 (hypothse nulle): Les moyennes de tous les groupes sont gales.
- H1 (hypothse alternative): Au moins une des moyennes des groupes est diffrente des autres.

### Conditions d'utilisation
- La variable quantitative est normalement distribue dans chaque groupe.
- Les variances des groupes sont gales.
- Les observations sont indpendantes.

### Interprtation
Si la p-valeur est infrieure au seuil de significativit (gnralement 0,05), on rejette l'hypothse nulle et on conclut qu'au moins une des moyennes des groupes est significativement diffrente des autres.


### Interprtation
Si la p-valeur est infrieure au seuil de significativit (gnralement 0,05), on rejette l'hypothse nulle et on conclut que les distributions des deux groupes sont significativement diffrentes.



```python
# ANOVA
anova_stat, anova_p_value = f_oneway(setosa, virginica, versicolor)
print("Statistique F de l'ANOVA :", anova_stat)
print("P-valeur (ANOVA) :", anova_p_value)
```

### Test de Levene pour l'galit des variances

Le test de Levene est utilis pour vrifier **l'hypothse d'galit des variances** entre plusieurs groupes. Il est particulirement utile aprs l'ANOVA pour vrifier l'hypothse d'homoscdasticit.

## Hypothses
- H0 (hypothse nulle): Les variances des diffrents groupes sont gales.
- H1 (hypothse alternative): Au moins une des variances des diffrents groupes est diffrente.

## Conditions d'utilisation
- Les chantillons sont indpendants.
- Les observations dans chaque groupe sont indpendantes et identiquement distribues.

## Interprtation
Si la p-valeur est infrieure  un seuil de signification prdfini (par exemple, 0,05), on rejette l'hypothse nulle et on conclut que les variances ne sont pas gales entre les groupes.

```python
from scipy.stats import levene

# Test de Levene pour l'galit des variances
levene_stat, levene_p_value = levene(setosa, virginica, versicolor)
print("Statistique W de Levene :", levene_stat)
print("P-valeur (Levene) :", levene_p_value)
```

## Test de Kruskal-Wallis

### Hypothses
- H0 (hypothse nulle): Les distributions de tous les groupes sont gales.
- H1 (hypothse alternative): Au moins une des distributions des groupes est diffrente des autres.

### Conditions d'utilisation
- Les observations sont indpendantes.
- Les donnes peuvent tre ordonnes (au moins ordinale).

```python
# Test de Kruskal-Wallis

kruskal_stat, kruskal_p_value = kruskal(setosa, virginica, versicolor)
print("Statistique H de Kruskal-Wallis :", kruskal_stat)
print("P-valeur (Kruskal-Wallis) :", kruskal_p_value)
```
